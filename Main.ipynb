```python
Y = df['Heart Attack Risk']
```

```python
X = df.drop('Heart Attack Risk', axis =1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 1)
```

### Model Training

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
rfclf = RandomForestClassifier()
```

```python
rfclf.fit(x_train, y_train)
```

```python
dtclf = DecisionTreeClassifier()
```

```python
dtclf.fit(x_train,y_train)
```

```python
lrclf = LogisticRegression(max_iter = 2000)
```

```python
lrclf.fit(x_train,y_train)
```

```python
gbclf = GradientBoostingClassifier()
```

```python
gbclf.fit(x_train,y_train)
```

```python
rf_accuracy = rfclf.score(x_test, y_test)
```

```python
dt_accuracy = dtclf.score(x_test, y_test)
```

```python
Ir_accuracy = lrclf.score(x_test, y_test)
```

```python
gb_accuracy = gbclf.score(x_test, y_test)
```

### Cross Validation

```python
rf_scores = cross_val_score(rfclf, X, Y, cv = 10)
mean_score1 = np.mean(rf_scores)
```

```python
dt_scores = cross_val_score(dtclf, X, Y, cv = 10)
mean_score2 = np.mean(dt_scores)
```

```python
Ir_scores = cross_val_score(lrclf, X, Y, cv = 10)
mean_score3 = np.mean(Ir_scores)
```

```python
gb_scores = cross_val_score(gbclf, X, Y, cv=10)
mean_score_gb = np.mean(gb_scores)
```

### Showing the accuracy

```python
print("The accuracy of random forest is ",rf_accuracy,"\nThe accuracy of DecisionTreeClassifier is ",dt_accuracy,
     "\nThe accuracy of LogisticRegression is ",Ir_accuracy,"\nThe accuracy of Gradient Boosting is", gb_accuracy,"\nThe accuracy of rfclf k_cross_validation is ",mean_score1,
     "\nThe accuracy of dtclf k_cross_validation is ",mean_score2,"\nThe accuracy of Irclf k_cross_validation is ",mean_score3,
      "\nThe accuracy of Gradient Boosting k_cross_validation is", mean_score_gb)
```
